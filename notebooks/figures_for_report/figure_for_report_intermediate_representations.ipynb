```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = np.array(Image.open("column_detection_screencapture_300x300px.png"))
```

```python
nm_x = 300*0.064
nm_y = nm_x#for 300px
```

```python
nm_x
```

```python
true_nm_per_px = len(img)/nm_x
labels = np.array([str(x) for x in np.arange(0,int(nm_x+1)*2)])
```

```python
ticks = np.arange(0,len(img), true_nm_per_px)
```

```python
fig, ax = plt.subplots(1, figsize=(4,4), dpi= 200)

ax.set_xticks(ticks[::2])
ax.set_yticks(ticks[::2])
ax.set_xticklabels(labels[::2], fontsize=8)
ax.set_yticklabels(labels[::2], fontsize=8)
ax.set_xlabel("µm")
ax.set_ylabel("µm")


ax.imshow(img)
```
